```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15,6


df = pd.read_csv('/home/lenovo-s145/Series_Temporais/Dados/italy_earthquakes_from_2016-08-24_to_2016-11-30.csv').set_index("Time")
df.head()
```

```python
df.index = pd.to_datetime(df.index)
df.head()
```

```python
df.describe()
```

```python
df['Magnitude'].resample('D').apply([np.mean]).plot()
plt.title('Magnetude média por dia')
plt.ylabel('Magnitude')
```

```python
df['Magnitude'].resample('2D').apply([np.mean]).plot()
plt.title('Magnetude média a cada 2 dias')
plt.ylabel('Magnitude')
```

```python
magn3 = df.loc[df['Magnitude'] >= 3.0, 'Magnitude']
pd.Series.rolling(magn3, window=25).mean().plot(style='-g')
plt.title("Média movel da magnetude")
plt.ylabel("Magnetude")
plt.xlabel("Tempo")
```
