```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
```

```python
data  = pd.read_csv('/home/lenovo-s145/Series_Temporais/Dados/avocado.csv')
```

```python
data.head()
```

```python
summer = data.groupby('Date')['Total Volume'].mean().reset_index()
```

```python
summer.head()
```

```python
fig, ax = plt.subplots(1,1, figsize=(15,8))
summer.set_index('Date').plot(ax=ax, marker='o', linestyle='--', color='blue')
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,10))
(summer.set_index("Date")
    .assign(month=lambda df: df.index.month)
    .groupby('month')['Total Volume'].agg(['mean','std','mediab','min','max'])
    .plot(ax=ax, marker='0', linewidth=3)
)
ax.set_xlabel('Ms')
```
